```python
#video file_path
filepath = "/content/Python_Tutorial__Logging_Basics_-_Logging_to_Files,_Setting_Levels,_and_Formatting(720p) (1).mp4"
```

```python
from moviepy.editor import VideoFileClip
import os

#function for extracting images from video
def video_to_images(video_path,output_folder):
  clip=VideoFileClip(video_path)
  clip.write_images_sequence(
      os.path.join(output_folder,"frame%04d.png"),fps=0.2
  )
```

```python

#function for extracting audiofile from video
def video_to_audio(video_path,output_audio_path):
  clip=VideoFileClip(video_path)
  audio=clip.audio
  audio.write_audiofile(output_audio_path)
```

```python
#function for converting audiofile to text

def audio_to_text(audio_path):
  recognizer=sr.Recognizer()
  audio=sr.AudioFile(audio_path)

  with audio as source:
    audio_data=recognizer.record(source)

    try:

      #recognize the speech
      text = recognizer.recognize_whisper(audio_data)

    except sr.UnknownValueError:
      print("Speech recognition could not understand the audio.")
  return text
```

```python
#extracting audio from video
video_to_audio(filepath,output_audio_path)
```

```python
#extracting images from video
video_to_images(filepath,output_folder)
```

```python
#converting audio to text
text_data=audio_to_text(output_audio_path)
```

```python
#converting text to document
with open(output_folder + "output_text.txt", "w") as file:
        file.write(text_data)
print("Text data saved to file")
file.close()
```

```python
#removing audio file since we have converted it into text
os.remove(output_audio_path)
print("Audio file removed")
```

```python
from google.colab import userdata
api_key = userdata.get('GROQ_API')
huggingface_api_key=userdata.get('HUGGING_FACE_API')

import os

# Make it available as an environment variable
os.environ["GROQ_API_KEY"] = api_key
os.environ["HUGGING_FACE_API"]=huggingface_api_key
```

```python
#constructing vectore store to store data

from llama_index.vector_stores.lancedb import LanceDBVectorStore
from llama_index.core import StorageContext
```

```python
text_store=LanceDBVectorStore(uri="lancedb",table_name="text_collection")
image_store=LanceDBVectorStore(uri="lancedb",table_name="image_collection")
storage_context=StorageContext.from_defaults(vector_store=text_store,image_store=image_store)
```

```python
output_folder
```

```python
from llama_index.core.indices import MultiModalVectorStoreIndex
from llama_index.core import SimpleDirectoryReader
from llama_index.core.node_parser import SentenceSplitter
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.groq import Groq
from llama_index.core import Settings

embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/clip-ViT-B-32")
llm = Groq(model="meta-llama/llama-4-maverick-17b-128e-instruct", api_key=api_key)


Settings.embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/clip-ViT-B-32")
Settings.llm= llm


```

```python
documents=SimpleDirectoryReader(output_folder).load_data()
text_splitter = SentenceSplitter(chunk_size=512, chunk_overlap=50)

index = MultiModalVectorStoreIndex.from_documents(
    documents,
    storage_context=storage_context,
    embed_model=embed_model,
    text_splitter=text_splitter,
)
```

```python
def plot_images(image_paths):
    images_shown = 0
    plt.figure(figsize=(16, 9))
    for img_path in image_paths:
        if os.path.isfile(img_path):
            image = Image.open(img_path)

            plt.subplot(2, 3, images_shown + 1)
            plt.imshow(image)
            plt.xticks([])
            plt.yticks([])

            images_shown += 1
            if images_shown >= 9:
                break
```

```python
test_query = "how do i use the logging function"
# generate  retrieval results
retriever = index.as_retriever(similarity_top_k=3, image_similarity_top_k=5)
retrieval_results = retriever.retrieve(test_query)
```

```python
from llama_index.core.response.notebook_utils import display_source_node
from llama_index.core.schema import ImageNode

retrieved_image = []
for res_node in retrieval_results:
    if isinstance(res_node.node, ImageNode):
        retrieved_image.append(res_node.node.metadata["file_path"])
    else:
        display_source_node(res_node, source_length=200)

plot_images(retrieved_image)
```

```python
from llama_index.core import PromptTemplate

# Main QA template
qa_template = PromptTemplate(
    "You are a multimodal assistant. "
    "Use both the text and image context to answer the question.\n\n"
    "Context:\n{context_str}\n\n"
    "Question:\n{query_str}\n\n"
    "Answer in detail, mentioning if the answer comes from text or image."
)

# Refinement template
refine_template = PromptTemplate(
    "We have an existing answer: {existing_answer}\n"
    "Here is more multimodal context:\n{context_msg}\n"
    "Please refine the answer if the new information adds value, "
    "otherwise keep it as is."
)
```

```python
query_engine = index.as_query_engine(
    text_qa_template=qa_template,
    refine_template=refine_template
)
```

```python
response = query_engine.query("what is the name of the tutor")
print(response)
```
